```python
import pandas as pd
import numpy as np
```

```python
col_names = ["empty","T-Day","T-Month","T-Year", "Code", "No", "R-Day","R-Month","R-Year",
         "Origin Court", "Origin Code", "Origin No", "Origin Year", "Specific Case Type",
         "Judge 1", "Judge 2", "Judge 3", "Judge 4", "Judge 5", "Judge 6", "Judge 7", 
         "Case coming for", "Case OutCome", "Reason of adjournment", "N-Day","N-Month",
         "N-Year","P-M", "P-F","P-Org.", "D-M", "D-F", "D-Org.", "Legal Rep", "Witness-P", "Witness-D",
         "Crim-Custody","Other Details"]
```

```python
len(col_names)
```

```python
sample_data = pd.read_excel("../Data Analysis/Data/Sample Data.xlsx", names = col_names)
sample_data.head(6)
```

```python
# dropping the header roles
sample_data.drop(index = [0,1,2,3], inplace = True)
```

```python
sample_data.head()
```

```python
sample_data.reset_index(drop = True, inplace = True)
```

```python
sample_data.head()
```

```python
# dropping the empty column
sample_data.drop(columns = ['empty'], inplace =True)
```

```python
sample_data.head()
```

```python
sample_data.columns
```

```python
sample_data.dtypes
```

```python
# create the columns todays Date
# if dtype is a number 
#~ covert  using .astype(str)
sample_data["Today's Date"] = sample_data["T-Day"].astype(str)+"-"+ sample_data["T-Month"]+"-"+ sample_data["T-Year"].astype(str)
sample_data.head()
```

```python
# convert Today's Date from string to a date type
sample_data["Today's Date"] = pd.to_datetime(sample_data["Today's Date"], format = '%d-%b-%Y')
```

```python
sample_data.dtypes
```

```python
# drop the columns that have been combined to a new one
sample_data.drop(columns = ['T-Day','T-Month','T-Year'], inplace = True)
```

```python
# move the Today's column to the first column
sample_data.insert(0, "Today's Date", sample_data.pop("Today's Date"))
```

```python
sample_data.head()
```

```python
# creating new column (Reg. Date) for concat R-(Day,Month,Year), and inserting in its correct location
sample_data.insert(3, "Reg. Date", sample_data["R-Day"].astype(str)+ "-"+sample_data["R-Month"]+ "-"+ sample_data["R-Year"].astype(str))
sample_data.head()
```

```python

# creating new column (Next activity Date) for concat R-(Day,Month,Year), and inserting in its correct location
sample_data.insert(23, "N. activity Date", sample_data["N-Day"].astype(str)+ "-"+sample_data["N-Month"]+ "-"+ sample_data["N-Year"].astype(str))
sample_data.head()
```

```python
sample_data.columns
```

```python
# converting 'Reg. Date' and 'N. activity Date' columns (string) to datetime
sample_data["Reg. Date"] = pd.to_datetime(sample_data["Reg. Date"], format = '%d-%b-%Y')
sample_data["N. activity Date"] = pd.to_datetime(sample_data["N. activity Date"], format = '%d-%b-%Y')
sample_data.head()
```

```python
# dropping unwanted columns
sample_data.drop(columns = ['R-Day','R-Month', 'R-Year','N-Day','N-Month', 'N-Year'], inplace =  True)
sample_data.head()
```
